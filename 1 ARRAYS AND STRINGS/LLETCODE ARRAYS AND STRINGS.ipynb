### 1. TWO SUM


```python
def TwoSum(arr,target):
    if len(arr)<2:
        return False
    
    seen = set()

    for i in range (len(arr)-1):
        complement = target - arr[i]
        if complement in seen:
            return [arr.index(complement),i]
        seen.add(arr[i])

TwoSum([2,7,11,15], 9)
```

## 26. Remove Duplicates from Sorted Array



```python
def RemoveDuplicates(nums):
    l= 1
    for r in range(l,len(nums)):
        if nums[r] != nums [r-1]:
            nums[l] = nums[r]
            l += 1

    return l

RemoveDuplicates([0,0,1,1,1,2,2,3,3,4])
```

## 27. Remove Element




```python
def removeElement(nums,val):
    l=0
    for r in range(len(nums)):
        if nums[r] != val:
            nums[l]=nums[r]
            l += 1

    return l


removeElement(nums = [0,1,2,2,3,0,4,2], val = 2)
```

## 66. Plus One


```python
def plus_one(digits):
    # Start from the last digit
    for i in range(len(digits) - 1, -1, -1):
        # If the current digit is not 9, just increment it and return
        if digits[i] < 9:
            digits[i] += 1
            return digits
        
        # If the current digit is 9, set it to 0
        digits[i] = 0

    # If all digits were 9, we need to add a leading 1
    return [1] + digits

# Example usage
digits = [9, 9, 3,2]
print("Result:", plus_one(digits))
```

## 67. Add Binary



```python
def add_binary(a: str, b: str) -> str:
    # Initialize variables
    i, j = len(a) - 1, len(b) - 1
    carry = 0
    result = []

    # Loop until both strings are processed
    while i >= 0 or j >= 0 or carry:
        # Get the current digit from each string or 0 if index is out of range
        digit_a = int(a[i]) if i >= 0 else 0
        digit_b = int(b[j]) if j >= 0 else 0
        
        # Calculate the sum of digits and the carry
        total = digit_a + digit_b + carry
        carry = total // 2  # Update carry for next iteration
        result.append(str(total % 2))  # Append the binary digit
        
        # Move to the next digits
        i -= 1
        j -= 1

    # Join the result in reverse order
    return ''.join(reversed(result))

# Example usage
a = "11001"
b = "1011"
print(add_binary(a, b))  # Output: "100"
```

## 125. Valid Palindrome



```python
def isPalindrome(s: str) -> bool:
    # Filter only alphanumeric characters and convert to lowercase
    filtered = ''.join(char.lower() for char in s if char.isalnum())
    # Check if the filtered string is equal to its reverse
    return filtered == filtered[::-1]

# Test Cases
print(isPalindrome("A man, a plan, a canal: Panama"))  # Output: True
print(isPalindrome("race a car"))  # Output: False
```

## 136. Single Number


```python
# SOLUTION ONE

def singleNumber(nums):
    result = 0
    for num in nums:
        result ^= num  # XOR accumulates the single number
    return result

# Example usage
nums = [4, 1, 2, 1, 2]
print("Single Number:", singleNumber(nums))
```

```python
# SOLUTION TWO
def singleNumber(nums):
    num_set = set()
    for num in nums:
        if num in num_set:
            num_set.remove(num)  # Remove if it's a duplicate
        else:
            num_set.add(num)  # Add if it's not in the set
    return num_set.pop()  # The remaining element is the single number

# Example usage
nums = [4, 1, 2, 1, 2]
print("Single Number:", singleNumber(nums))
```

## 169. Majority Element


```python
# Optimal Approach: Boyer-Moore Voting Algorithm

def majorityElement(nums):
    candidate, count = None, 0

    # Finding the candidate
    for num in nums:
        if count == 0:
            candidate = num
        count += (1 if num == candidate else -1)

    # The candidate is guaranteed to be the majority element.
    return candidate

# Example usage
nums = [2, 2, 1, 1, 1, 2, 2]
print("Majority Element:", majorityElement(nums))
```

## 344. Reverse String


```python
def reverseString(s):
    left,right = 0, len(s)-1
    while left < right:
        #swap elements at left and right indices
        s[left],s[right] = s[right],s[left]

        left += 1
        right -= 1

    return s



reverseString(s = ["h","e","l","l","o"])
```

## 704. Binary Search



```python
def search(nums,target):

    left,right = 0, len(nums)-1

    while left <= right:
        mid = (left + right)//2
        if nums[mid]== target:
            return mid

        elif nums[mid]< target:
            left = mid + 1

        else: 
            right = mid -1

    return -1

    
    
search(nums = [-1,0,3,5,9,12], target = 2)
```

## 905. Sort Array By Parity


```python
# Leetcode 905: Sort Array By Parity

from typing import List

def sort_array_by_parity_two_pointers(nums: List[int]) -> List[int]:
    left, right = 0, len(nums) - 1
    while left < right:
        if nums[left] % 2 > nums[right] % 2:
            nums[left], nums[right] = nums[right], nums[left]
        if nums[left] % 2 == 0:
            left += 1
        if nums[right] % 2 == 1:
            right -= 1
    return nums


def sort_array_by_parity_comprehension(nums: List[int]) -> List[int]:
    return [x for x in nums if x % 2 == 0] + [x for x in nums if x % 2 == 1]


# Test cases
nums = [3, 1, 2, 4]
print("Two Pointers:", sort_array_by_parity_two_pointers(nums))
print("List Comprehension:", sort_array_by_parity_comprehension(nums))
```

## 922. Sort Array By Parity II


```python
def sortArrayByParityII(nums):
    i, j = 0, 1
    n = len(nums)

    while i < n and j < n:
        # Find the misplaced even element
        while i < n and nums[i] % 2 == 0:
            i += 2
        # Find the misplaced odd element
        while j < n and nums[j] % 2 == 1:
            j += 2

        # If both are within bounds, swap them
        if i < n and j < n:
            nums[i], nums[j] = nums[j], nums[i]

    return nums

# Example usage
nums = [4, 2, 5, 7]
print("Sorted by parity:", sortArrayByParityII(nums))
```

## 1299. Replace Elements with Greatest Element on Right Side


```python
def replace_elements(arr):
    n = len(arr)
    max_right = -1  # Start with -1 as the last element should be -1

    # Traverse the array from right to left
    for i in range(n - 1, -1, -1):
        # Store the current value before changing it
        current = arr[i]
        # Replace with the max value encountered so far
        arr[i] = max_right
        # Update max_right for the next iteration
        max_right = max(max_right, current)

    return arr

# Example usage
arr = [17, 18, 5, 4, 6, 1]
print(replace_elements(arr))
```

## 1365. How Many Numbers Are Smaller Than the Current Number


```python
# SOLUTION ONE: BRUTE FORCE (TIME COMPLEXITY (O(n^2)))
def smaller_numbers_than_current(nums):
    result = []
    for i in range(len(nums)):
        count = sum(1 for j in nums if j < nums[i])
        result.append(count)
    return result

# Example usage
nums = [8, 1, 2, 2, 3]
print(smaller_numbers_than_current(nums))  # Output: [4, 0, 1, 1, 3]
```

```python
# SOLUTION TWO: SORTING (TIME COMPLEXITY: O(n log n))
def smaller_numbers_than_current(nums):
    sorted_nums = sorted(nums)
    rank = {}
    
    for i, num in enumerate(sorted_nums):
        if num not in rank:
            rank[num] = i
            
    return [rank[num] for num in nums]

# Example usage
nums = [8, 1, 2, 2, 3]
print(smaller_numbers_than_current(nums))  # Output: [4, 0, 1, 1, 3]
```

## 242. Valid Anagram


```python
# SOLUTION ONE: SORTING   Time Complexity O(n log n)
def is_anagram(s: str, t: str) -> bool:
    return sorted(s) == sorted(t)

# Test cases
print(is_anagram("anagram", "nagaram"))  # Output: True
print(is_anagram("rat", "car"))          # Output: False
```

```python
#Approach 2: Using Hash Map (Optimal)  Time Complexity: O(n)
from collections import Counter

def is_anagram(s: str, t: str) -> bool:
    return Counter(s) == Counter(t)

# Test cases
print(is_anagram("anagram", "nagaram"))  # Output: True
print(is_anagram("rat", "car"))          # Output: False
```

```python
# Alternative without Counter
def is_anagram(s: str, t: str) -> bool:
    if len(s) != len(t):
        return False
    
    count = {}
    
    for char in s:
        count[char] = count.get(char, 0) + 1
        
    for char in t:
        if char not in count or count[char] == 0:
            return False
        count[char] -= 1
        
    return all(value == 0 for value in count.values())

# Test cases
print(is_anagram("anagram", "nagaram"))  # Output: True
print(is_anagram("rat", "car"))          # Output: False
```

## 3 Longest Substring Without Repeating Characters



```python
def longestSubstring(s):
    sett = set()
    l=0

    sett = set()
    l = 0
    w = 0

    for r in range(len(s)):
        while s[r] in sett:
            sett.remove(s[l])
            l += 1
        sett.add(s[r])
        w = max(w, r - l + 1)

    return w

longestSubstring(s = "pwwkew")
```

## 15 3Sum


```python
def threeSum(nums):
    res = []
    nums.sort()

    for i,a in enumerate (nums):
        if i > 0 and a == nums [i-1]:
            continue

        l,r = i+1, len(nums)-1
        while l<r:
            threeSum = a + nums[l] + nums[r]
            if threeSum > 0:
                r -= 1
            elif threeSum < 0:
                l += 1
            else:
                res.append([a,nums[l],nums[r]])
                l += 1
                while nums[l] == nums[l-1] and l<r:
                    l += 1

    return res

threeSum(nums = [-1,0,1,2,-1,-4])
```
