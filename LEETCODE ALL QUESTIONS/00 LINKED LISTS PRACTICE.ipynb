```python
class ListNode():
    def __init__(self,data,next=None):
        self.data = data
        self.next = next

class LinkedList():
    def __init__(self):
        self.head =None

    def append(self,data):
        new_node = ListNode(data)

        if not self.head:
            self.head = new_node
            return

        curr = self.head
        while curr.next:
            curr = curr.next

        curr.next = new_node
    def insert_beginning(self,data):
        new_node = ListNode(data)
        if self.head is None:
            self.head = new_node
            return        
        new_node.next=self.head
        self.head = new_node


    def insert_pos(self,data,pos):

        new_node = ListNode(data)

        curr= self.head
        
        for _ in range(pos-1):
            curr =curr.next

        new_node.next=curr.next
        curr.next =new_node

    def delete_last(self):

        curr= self.head
        while curr.next.next:
            curr = curr.next

        curr.next =None


    def delete_first(self):
        self.head = self.head.next



    def delete_node(self,key):

        if self.head.data == key:
            self.head =self.head.next
            return
        
        curr = self.head
        while curr.next and curr.next.data !=key:
            curr = curr.next

        curr.next = curr.next.next



    def reverse(self):

        prev = None
        

        curr = self.head
        while curr:
            nxt = curr.next
            curr.next = prev
            prev= curr
            curr = nxt

        self.head = prev
            






    def traverse(self):
        if not self.head:
            return "Empty List"
        
        curr = self.head
        while curr:
            print (str(curr.data), end= ' -> ')
            curr = curr.next

        print("None")


    def printf(self):
        print(self.head.data)
        


sll = LinkedList()
sll.append(1)
sll.append(2)
sll.append(4)
sll.append(5)
sll.insert_beginning(0)
sll.insert_pos(3,3)
#sll.delete_last()
#sll.delete_first()
sll.delete_node(0)
sll.traverse()

sll.reverse()
sll.traverse()

#sll.printf()


        
```

```python
# MERGING TWO LINKED LISTS
l1 = LinkedList()
l1.append(1)
l1.append(3)
l1.append(5)


l2 = LinkedList()
l2.append(2)
l2.append(4)
l2.append(6)





def merge(list1,list2):
    dummy = ListNode(0)
    tail =dummy

    l1 = list1.head
    l2 = list2.head
    while l1 and l2:
        if l1.data < l2.data:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next

    # Attach the remaining nodes
    if l1:
        tail.next = l1
    else:
        tail.next = l2

    # Create a new LinkedList to return
    merged_list = LinkedList()
    merged_list.head = dummy.next
    return merged_list

merged = merge(l1,l2)
merged.traverse()
```
