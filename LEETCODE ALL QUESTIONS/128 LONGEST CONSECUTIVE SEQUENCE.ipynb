```python
def longestConsecutive(nums):
    if not nums:
        return 0

    num_set = set(nums)
    longest = 0

    for num in num_set:
        # only start counting when num is the start of a sequence
        if num - 1 not in num_set:
            current = num
            curr_max = 1

            while current + 1 in num_set:
                current += 1
                curr_max += 1

            longest = max(longest, curr_max)

    return longest


longestConsecutive(nums=[100,4,200,1,3,2]
)
```
